```python
def recommend(movie_name):
    index=np.where(pt.index==movie_name)[0][0]     #index of each movie
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

    for i in similar_items:
        print(pt.index[i[0]])
```

```python
average_pt=pt.mean(axis=1)
```

```python
average_pt
```

```python
def predict_rating(u,movie_name):
  neu_i=float(average_pt[movie_name])
  mvs= pt.index.tolist()
  i=np.where(pt.index==movie_name)[0][0]
  sum=0
  sum2=0
  k=5
  similarities = similarity_score[i]
  top_k_indices = np.argsort(similarities)[-k-1:-1] #top k similarities
  for j in top_k_indices:
    mv=mvs[j]
    sim=similarity_score[i][j]
    rt=pt.at[mv,u]
    if rt==0:
      continue
    sum=sum+sim*(rt-float(average_pt[mv]))
    sum2=sum2+sim
  if sum2==0:
    return neu_i
  else:
    return neu_i+sum/sum2
```

```python
print(float(predict_rating(20,"(500) Days of Summer")))
```

```python
import numpy as np
import random

def evaluate_model_with_sampling(sample_size=10000):
    actual_ratings = []
    predicted_ratings = []

    # Collect all possible non-zero (movie, user) pairs
    non_zero_pairs = [
        (movie, user) for movie in pt.index for user in pt.columns if pt.at[movie, user] != 0
    ]

    # Randomly sample a subset of pairs
    sampled_pairs = random.sample(non_zero_pairs, min(sample_size, len(non_zero_pairs)))

    # Process sampled movie-user pairs
    for movie, user in sampled_pairs:
        actual_ratings.append(pt.at[movie, user])
        predicted_ratings.append(predict_rating(user, movie))

    # Convert to numpy arrays
    actual_ratings = np.array(actual_ratings)
    predicted_ratings = np.array(predicted_ratings)

    # Compute MAE
    mae = np.mean(np.abs(actual_ratings - predicted_ratings))
    return mae

# Run evaluation on a sample of 10,000 movie-user pairs
mae_sampled = evaluate_model_with_sampling(sample_size=10000)
print(f"Sampled MAE Score: {mae_sampled:.4f}")
```
